```python
import graphlab
```

```python
image_train = graphlab.SFrame('image_train_data/')
```

```python
image_train.head()
```

## Train a Nearest-Neighbors Model

```python
knn_model = graphlab.nearest_neighbors.create(image_train,
                                             features=['deep_features'],
                                             label='id')
```

```python
#
# Using the model to find similar images

graphlab.canvas.set_target('ipynb')
cat = image_train[18:19]
cat['image'].show()
```

```python
knn_model.query(cat)
```

```python
#
# Getting the images by the ID

def get_images(query_result):
    return image_train.filter_by(query_result['reference_label'], 'id')
```

```python
cat_neighbors = get_images(knn_model.query(cat))
```

```python
cat_neighbors['image'].show()
```

```python
car = image_train[8:9]
car['image'].show()
```

```python
 get_images(knn_model.query(car))['image'].show()
```

```python
#
# Creating a lambda to show the nearest neighbors! 

show_neighbors = lambda i:get_images(knn_model.query(image_train[i:i+1]))
```

```python
show_neighbors(8)
```

```python
show_neighbors(12)
```

## Creating Categories

```python
dog_train = image_train[image_train['label']=='dog']
cat_train = image_train[image_train['label']=='cat']
automobile_train = image_train[image_train['label']=='automobile']
bird_train = image_train[image_train['label']=='bird']
```

## Finding the nearest neighbor of each new category

```python
#
# Getting the test data

image_test = graphlab.SFrame('image_test_data/')
```

```python
# 
# DOG

knn_model_dog = graphlab.nearest_neighbors.create(dog_train,
                                             features=['deep_features'],
                                             label='id')
```

```python
#
# Using the model to get the neighbirs from the new category just created

dog_neighbors = get_images(knn_model_dog.query(dog_train))
```

```python
dog_neighbors['image'].show()
```

```python
# 
# CAT

knn_model_cat = graphlab.nearest_neighbors.create(cat_train,
                                             features=['deep_features'],
                                             label='id')
```

```python
#
# Using the model to get the neighbirs from the new category just created

cat_neighbors = get_images(knn_model_cat.query(cat_train))
```

```python
cat_neighbors['image'].show()
```
